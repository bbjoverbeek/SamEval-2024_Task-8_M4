```python
import json
import statistics
```

```python
with open("features/SubtaskA/train_monolingual/sentences.json", "r") as f:
    sentences = json.load(f)
```

```python
lengths = map(lambda x: (x[0], len(x[1])), sentences.items())
lengths = list(lengths)
```

```python
sorted(lengths, key=lambda x: x[1], reverse=True)
```

```python
import statistics
print("mean", statistics.mean(map(lambda x: x[1], lengths)))
print("median", statistics.median(map(lambda x: x[1], lengths)))
```
